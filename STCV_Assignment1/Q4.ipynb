```python
rows=1024
columns=1024
```

```python
for r in range(rows):
    print(grayImage[r]," ")
```

```python
pixels=[]
for r in range(rows):
    for c in range(columns):
        pixels.append(grayImage[r][c])
```

```python
pixDict={}
for i in pixels:
    if i in pixDict.keys():
        pixDict[i]+=1
    else:
        pixDict[i]=1
```

```python
pixDict
```

```python
countPixels=[0 for i in range(256)]
```

```python
for i in pixels:
    countPixels[i]+=1
```

```python
len(countPixels)
```

```python
numOfPixels=len(pixels)
```

```python
ind=-1
variance=float('inf')
currSum=0
for i in range(len(countPixels)):
    currSum+=countPixels[i]
    W1=currSum/numOfPixels
    W2=(numOfPixels-currSum)/numOfPixels
    V1=0
    V2=0
    mean1=0
    mean2=0
    s1=0
    s2=0
    for j in range(i+1):
        mean1+=countPixels[j]*j
        s1+=countPixels[j]
    for j in range(i+1,len(countPixels)):
        mean2+=countPixels[j]*j
        s2+=countPixels[j]
    if s1!=0:
        mean1=mean1/s1
    if s2!=0:
        mean2=mean2/s2
    
    for j in range(i+1):
        V1+=pow(j-mean1,2)*countPixels[j]
    for j in range(i+1,len(countPixels)):
        V2+=pow(j-mean2,2)*countPixels[j]
    if s1!=0:
        V1=V1/s1
    if s2!=0:
        V2=V2/s2
    
    var=(W1*V1)+(W2*V2)
    if var<variance:
        variance=var
        ind=i
```

```python
print(ind,variance)
```

```python
resultArr=[[0 for i in range(columns)] for j in range(rows)]
for r in range(rows):
    for c in range(columns):
        if grayImage[r][c]>ind:
            resultArr[r][c]=[255,255,255]
        else:
            resultArr[r][c]=[0,0,0]
```

```python
resultImage=np.asarray(resultArr)
io.imshow(resultImage)
```
