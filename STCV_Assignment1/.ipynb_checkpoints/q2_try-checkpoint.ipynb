```python
cars=[]
for i in range(1,12):
    name=str(i)+'.jpg'
    img=Image.open('Q2_train/Car/'+name)
    reimg =np.asarray(img.resize((256,256)))
    grayImg=cv2.cvtColor(reimg, cv2.COLOR_BGR2GRAY )
    fd, hog_image = hog(grayImg, orientations=9, pixels_per_cell=(16, 16),cells_per_block=(2, 2), visualize=True, multichannel=False)
    cars.append(fd)
```

```python
dogs=[]
for i in range(1,9):
    name=str(i)+'.jpg'
    img=Image.open('Q2_train/Dogs/'+name)
    reimg =np.asarray(img.resize((256,256)))
    grayImg=cv2.cvtColor(reimg, cv2.COLOR_BGR2GRAY )
    fd, hog_image = hog(grayImg, orientations=9, pixels_per_cell=(16, 16),cells_per_block=(2, 2), visualize=True, multichannel=False)
    dogs.append(fd)
```

```python
data={}
features=[]
label=[]
for ind in range(len(cars)):
    i=cars[ind]
    features.append(i)
    label.append(0)
for ind in range(len(dogs)):
    i=dogs[ind]
    features.append(i)
    label.append(1)
```

```python
data['Feature']=features
data['label']=label
```

```python
dataframe=pd.DataFrame(data)
```

```python
dataframe
```

```python
from sklearn.model_selection import train_test_split

```

```python
X_train, X_test, y_train, y_test = train_test_split(dataframe['Feature'], dataframe['label'], test_size=0.30, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
list(X_train)
```

```python
knn.fit(list(X_train),y_train)
```

```python
predictKnn=knn.predict(list(X_test))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictKnn))
```

```python
print(confusion_matrix(y_test,predictKnn))
```
