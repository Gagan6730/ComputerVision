```python
import cv2
import numpy as np
from skimage import io
```

```python
img = cv2.imread('kungfu.png',0)
img = cv2.medianBlur(img,5)
copyImg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
```

```python
laplacian = cv2.Laplacian(img,0)
```

```python
io.imshow(laplacian)
```

```python
circles = cv2.HoughCircles(laplacian,cv2.HOUGH_GRADIENT,1,50,
                            param1=200,param2=50,minRadius=0,maxRadius=0)

# circles=cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1.2, 100)
# circles=cv2.HoughCircles(img, cv2.HOUGH_GRADIENT,1, 450, 200, 100, 0, 0)
```

```python
circles
```

```python
circles = np.uint16(np.around(circles))
for x,y,r in circles[0,:]:
    # draw the outer circle
    cv2.circle(copyImg,(x,y),r,(255,0,0),2)
    # draw the center of the circle
    cv2.circle(copyImg,(x,y),2,(0,0,255),3)
```

```python
io.imshow(copyImg)
```
